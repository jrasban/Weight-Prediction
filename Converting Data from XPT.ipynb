```python
import pandas as pd
```

# Combine and Select Features 

```python
bmx = pd.read_sas('BMX_J.XPT')
demo = pd.read_sas('DEMO_J.XPT')
# print(data)
# print(data2)
# print(data3)
# print(data4)
demo.set_index('SEQN', inplace=True)
bmx.set_index('SEQN', inplace=True)

# frames = [demo, bmx]
combdata = demo.join(bmx, on=['SEQN'])
list = ['RIAGENDR'
        ,'RIDAGEYR'
        ,'RIDRETH3'
        ,'DMQMILIZ'
        ,'DMDCITZN'
        ,'DMDMARTL'
        ,'INDHHIN2'
        ,'BMXWT'
        ,'BMXHEAD'
        ,'BMXHT'
        ,'BMXBMI'
        ,'BMXLEG'
        ,'BMXARML'
        ,'BMXARMC'
        ,'BMXWAIST'
        ,'BMXHIP'
        ,'DMDBORN4'
        ,'DMDEDUC2']

combdata = combdata.loc[:, list]
combdata = combdata.rename(columns={'RIAGENDR':'Gender'
                                    ,'RIDAGEYR':'Age'
                                    ,'RIDRETH3':'Race'
                                    ,'INDHHIN2':'Annual HH Income'
                                    ,'BMXWT':'Weight'
                                    ,'BMXHEAD':'Head Circumference'
                                    ,'BMXHT':'Height'
                                    ,'BMXBMI':'BMI'
                                    ,'BMXLEG':'Upper Leg Length'
                                    ,'BMXARML':'Upper Arm Length'
                                    ,'BMXARMC':'Arm Circumference'
                                    ,'BMXWAIST':'Waist Circumference'
                                    ,'BMXHIP':'Hip Circumference'
                                    })
print(combdata)
combdata.to_csv('data.csv')
```
