```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data.csv')
data.set_index('SEQN',inplace=True)
data.columns
```

```python
keep = ['Gender', 'Age', 'Race','Weight', 'Height', 'BMI',
       'Upper Leg Length', 'Upper Arm Length', 'Arm Circumference',
       'Waist Circumference', 'Hip Circumference']
```

```python
data = data[keep]
```

```python
data['Race'] = data['Race'].replace({1:'Mexican American',2:'Other Hispanic',3:'White', 4:'Black',6:'Asian',7:'Other'})
data['Gender'] = data['Gender'].replace({1:'Male',2:'Female'})
filter1= data['Age']>=18

data = data[filter1]
```
